The gradient is the vector of partial derivatives of a function. It points in the direction of steepest increase, and its magnitude tells you how fast the function increases in that direction. The gradient is zero at critical points, which may be local minima, maxima, or saddle points.

This code requires pytorch to track operations involving x so later it can compute gradients with respect to x.

```python
import torch 

x= torch.randn(3, requires_grad=True)
print(x)
y= x+2
print(y)
z = y*y*2
z = z.mean()
print(z)

z.backward() #dz/dx
print(x.grad)
```

When a function is composed of multiple functions (like in a neural network), the chain rule is used to compute the gradient of the overall composition by combining the gradients (partial derivatives) of each function in the chain.

The Chain Rule is a way to find the derivative of composite functions.It states that the derivative of composite function f(g(x)) is f'(g(x)) g'(x). 


Stochastic Gradient Descent (SGD) is an optimization method that addresses the inefficiency of using the entire dataset for gradient computation in large-scale learning.
Instead of calculating gradients on the full dataset (as in batch gradient descent), SGD updates the model using only one random data point (or a small batch) per iteration.
