An activation function is applied to the output of each neuron to introduce non-linearity into the network. This is essential because, without it, even a deep neural network would collapse into a simple linear model, unable to capture complex patterns.

Common activation functions include:

ReLU (Rectified Linear Unit): Outputs 0 for negative inputs, and the input itself if positive. Fast and effective, especially in hidden layers.

Sigmoid: Squashes output between 0 and 1; often used in binary classification tasks.

Tanh: Similar to sigmoid but outputs between -1 and 1.

Softmax: Used in the output layer of classification networks with multiple classes.


Non-linearity allows the network to model complex functions, learn interactions between features, and make decisions based on hierarchical representations. In simple terms, it allows the network to "think beyond a straight line."

ReLU (Rectified Linear Unit) activation is defined by 
A(x)=max(0,x), this means that if the input x is positive, ReLU returns x, if the input is negative, it returns 0.

Key Advantages of ReLU

Simplicity and Efficiency: ReLU is computationally inexpensive as it involves only a thresholding operation. This is crucial for training deep networks with millions of parameters.

Non-Linearity: Despite its simplicity, ReLU introduces non-linearity, enabling neural networks to learn complex patterns.

Sparse Activation: ReLU outputs zero for negative inputs, leading to sparse activations where only a fraction of neurons are active at a time. This improves computational efficiency and reduces overfitting.

Avoids Vanishing Gradients: Unlike sigmoid or tanh functions, ReLU does not saturate for positive inputs, which helps mitigate the vanishing gradient problem and allows gradients to flow effectively during backpropagation.

Limitations of ReLU

Dying ReLU Problem: Neurons can become inactive (outputting zero) for all inputs if they receive large negative gradients during training.

Unbounded Output: ReLU outputs are unbounded on the positive side, which can sometimes lead to exploding gradients

Non-Differentiability at Zero: The derivative of ReLU is undefined at x = 0, though in practice, this is not a significant issue.

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

# Define ReLU
relu = nn.ReLU()

# Generate input data
x = torch.linspace(-10, 10, 100)
y = relu(x)

# Plot ReLU function
plt.plot(x.numpy(), y.numpy())
plt.title("ReLU Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.show()

# Apply ReLU to a tensor
tensor = torch.tensor([[-2.0, 0.0, 3.0], [1.0, -1.0, 2.0]])
activated = relu(tensor)
print("Input:\n", tensor)
print("Output after ReLU:\n", activated)
```
